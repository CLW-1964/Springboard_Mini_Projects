Interesting, the missing values in the name field are not NaN's, so an additional step is needed in order to backfill

```python
# Step 1 sort the df
jn_sort = jn.sort_values(['code', 'name'])
jn_sort.head(10)
```

```python
# load numpy in order to quickly replace the blanks with NaN's
import numpy as np

jn_sort.name[jn_sort.name == ""] = np.nan
jn_sort.head(10)
```

```python
# use bfill to replace the NaN's
jn_final = jn_sort.fillna(method='bfill')
jn_final.head(10)
```
